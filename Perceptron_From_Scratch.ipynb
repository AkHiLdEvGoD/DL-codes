# Coding a **Basic Perceptron Model** from scratch

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

## Making a Dummy Classification Dataset

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=50)
```

```python
def step(z):
    return 1 if z>0 else 0
```

## Main Function

- Added a bias column = 1 in the original X dataset
- Initialized a weight array with 1's and same no. of cols as X
- Ran a loop no. epochs time and selected a random row no.
- Updated the weight array according to the rule where we shift the random line towards the misclassified point by adjusting the line's coefficients.

```python
def perceptron(X,y,lr=0.1,epochs=200):
    X = np.insert(X,0,1,axis=1)
    weight = np.ones(X[0].shape)
    
    for _ in range(epochs):
        row = np.random.randint(0,100)
        y_hat = step(np.dot(X[row],weight))
        weight = weight + lr*(y[row]-y_hat)*X[row]
    
    return weight[0],weight[1:]
```

```python
bias_,weights_ = perceptron(X,y,0.1,300)
```

```python
print(bias_)
print(weights_)
```

```python
m = -(weights_[0]/weights_[1])
intercept = -(bias_/weights_[1])
```

## Plotting the best fit line obtained from our perceptron model

```python
x_line = np.linspace(-3,3,100)
y_line = (m * x_line) +intercept
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=50)
plt.plot(x_line,y_line)
plt.ylim(-3,2)
```
